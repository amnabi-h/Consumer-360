```python
import pandas as pd
import sqlite3

# 1. Load your Excel data
df = pd.read_excel("./01_Sales Dataset.xlsx")

# 2. Clean the column names (so SQL doesn't get confused by the '\n')
df.columns = df.columns.str.strip().str.replace('\n', '', regex=False).str.replace(' ', '_')

# 3. Create a real SQL Database file in your folder
conn = sqlite3.connect('Consumer360_Database.db')

# 4. Push the data into a table called 'sales_table'
df.to_sql('sales_table', conn, if_exists='replace', index=False)

print("✅ Success! Your Excel data is now inside a SQL Database file.")
```

```python
# Function to run SQL and show results
def run_sql(query):
    return pd.read_sql_query(query, conn)

# QUERY 1: Show me the top 5 most profitable cities
query1 = """
SELECT City, SUM(Profit) as Total_Profit
FROM sales_table
GROUP BY City
ORDER BY Total_Profit DESC
LIMIT 5;
"""
print("--- TOP 5 CITIES BY PROFIT ---")
display(run_sql(query1))
```

```python
# QUERY 2: Which categories sold more than 500 items?
query2 = """
SELECT Category, SUM(Quantity) as Items_Sold
FROM sales_table
GROUP BY Category
HAVING Items_Sold > 500;
"""
print("\n--- HIGH VOLUME CATEGORIES ---")
display(run_sql(query2))
```
